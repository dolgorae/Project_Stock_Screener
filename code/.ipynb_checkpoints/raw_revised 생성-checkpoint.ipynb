```python
import os
import pandas as pd
```

```python
group_price_df = pd.read_csv("group_price_df.csv")

k = group_price_df["종목코드"]
code = []
for i in k:
    if len(str(i)) < 6:
        a = (6 - len(str(i))) * '0' + str(i)
        code.append(a)
    else:
        code.append(str(i))



group_price_df = group_price_df.drop("Unnamed: 0", axis=1)
group_price_df["종목코드"] = code
group_price_df = group_price_df.set_index("종목코드")
```

```python
for i in code:
    data = pd.read_csv(os.getcwd() + "\\재무\\%s_naverfinance.csv" %i, encoding = 'cp949')
    data = data.drop(["Unnamed: 0", "연간.5", "연간.6", "연간.7"], axis = 1)
    data.loc[34] = ["GP"] + group_price_df.loc[i].tolist()
    data.to_csv("재무revised\\%s.csv" %i, encoding = 'cp949')
```
