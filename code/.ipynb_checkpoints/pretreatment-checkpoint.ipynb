```python
import os
import pandas as pd
```

```python
corplist = pd.read_csv("C:/Users/eaststate/Desktop/스마/Project_Stock_Screener/KOSPI200.csv",  encoding = 'cp949')
```

```python
corplist = corplist['종목']
```

```python
if not os.path.isdir("by_year_data"):
    os.makedirs("by_year_data")
```

```python
for i in ['balance', 'income', 'cash']:
    for corp in corplist:
        if os.path.isfile("재무제표\\%s_%s.csv" %(corp, i)):
            print(True)
            df = pd.read_csv("재무제표\\%s_%s.csv" %(corp, i), encoding='cp949')
            
            df = df.set_index("Unnamed: 0")
            column_list = df.columns.tolist()
            column_list.remove("class1")
            column_list.remove("class2")
            a= column_list.copy()
            b= column_list.copy()
            del a[-1]
            b.remove("2019")
            for pre_year, post_year in zip(b, a):
                df[post_year] = df[post_year] - df[pre_year]
                df.rename(columns = {post_year : post_year + "to" + pre_year}, inplace = True)
            df = df.drop([column_list[-1]], axis = 1)
        
            
            df.to_csv("by_year_data\\%s_%s_by_year.csv" %(corp,i), encoding='utf-8-sig')
            print("Download %s_%s complete" %(corp, i))
            
        else:
            print("There is no file name")

        
```
